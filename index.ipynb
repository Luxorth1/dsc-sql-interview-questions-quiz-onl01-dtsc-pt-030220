## Question 2

Explain the relationship between **Primary Keys** and **Foreign Keys**.

Primary Keys are unique values within a table used as identifiers, a foreign key is a column that acts as the Primary key for another table. This allows for alot of relational DB stuff to occur, such as simple joins.




## Question 3

Explain the different types of relationships entities can have in a SQL database. 

Type your answer below this line:
_______________________________________________________________________________________________________________________________
One-to-One: These are when one item in a table directly ties to one other item in another table.
One-to-Many: These are when one item can tie to multiple different items, for example in the Data.sqlite example we had 1 product, however this can tie to many different order details, but each order detail ties back to a product.
Many-to-Many: These are when a table has multiple values which can be tied to multiple values on another table, such as if we had an employee who works from 3 different offices, tying their employee row to the offices table we will retreive 3 copies of the row with different offices if queried.



## Question 4

Explain the various types of JOINs possible with SQL. 

Type your answer below this line:
_______________________________________________________________________________________________________________________________
Left: This is where it keeps all values on the left column and only brings the values that match from the right
Right: This is the opposite of the left join, it returns all values that match directly to the table on the right, does not return values that do not match.
Inner: Drops all values that do not match, i.e. if there are not direct relationships between rows it does not display them
Outer: This just puts both tables together regardless of matches with keys, it ties together values that do match, but also inserts rows that may be missing data if there are no matches.
